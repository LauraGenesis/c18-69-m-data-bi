```python
datos=pd.read_csv("Casos_dengue.csv", index_col=0)
```

```python

# Nuevo nombre para las columnas
new_column_names = {
    'ID': 'Id',
    'Año': 'Año',
    'Semana Epidemiológica (a)': 'Semana_Epidemiologica',
    'Total de Casos de Dengue (b)': 'Total_Casos',
    'Tasa de Incidencia (c)': 'Tasas_Incidencias',
    'Confirmados Laboratorio': 'Confirmados_Laboratorio',
    '% Lab Conf (x100)': '%_Laboratorio_Confirmado',
    'Dengue Grave (d)': 'Dengue_Grave',
    '(DG/D) x100 (e)': '%_Dengue_Grave',
    'Muertes': 'Muertes',
    'Letalidad (f)': 'Letalidad',
    'Población X 1000': 'Poblacion_x_1000',
    'ID_Pais_Subregion': 'ID_Pais',
    'ID_Serotipo': 'ID_Serotipo',
    'ID_Calendario': 'ID_Calendario'
}

# Renombrar columnas usando rename()
datos.rename(columns=new_column_names, inplace=True)

```

```python
datos
```

```python
datos.columns
```

```python
# Eliminar los separadores de miles y convertir comas decimales a puntos
columns_to_convert = ['Tasas_Incidencias', '%_Laboratorio_Confirmado','%_Dengue_Grave']
for column in columns_to_convert:
    datos[column] = datos[column].str.replace('.', '', regex=False)  # Eliminar separadores de miles
    datos[column] = datos[column].str.replace(',', '.', regex=False)  # Reemplazar comas decimales por puntos
    datos[column] = datos[column].astype(float)  # Convertir a float

# Redondear la columna 'Tasas_Incidencias' a 2 decimales y convertir a string
datos['Tasas_Incidencias'] = datos['Tasas_Incidencias'].round(2).astype(str)

# Mostrar el DataFrame para verificar los cambios
datos
```

```python
datos[['Año', 'Semana_Epidemiologica', 'Total_Casos', 'Tasas_Incidencias',
       'Confirmados_Laboratorio', '%_Laboratorio_Confirmado', 'Dengue_Grave',
       '%_Dengue_Grave', 'Muertes', 'Letalidad', 'Poblacion_x_1000', 'ID_Pais',
       'ID_Serotipo', 'ID_Calendario']]
```

```python
datos[['Tasas_Incidencias',"%_Laboratorio_Confirmado","%_Dengue_Grave"]] = datos[['Tasas_Incidencias',"%_Laboratorio_Confirmado","%_Dengue_Grave"]].round(2).astype(str)
datos
```

```python
# Columnas que queremos convertir a int
columns_to_convert = ['Confirmados_Laboratorio', 'Dengue_Grave', 'Semana_Epidemiologica', 'Total_Casos', 'Muertes', 'Letalidad']

# Reemplazar comas decimales por puntos y eliminar separadores de miles
for column in columns_to_convert:
    datos[column] = datos[column].replace('.', '', regex=False)  # Eliminar separadores de miles
    datos[column] = datos[column].replace(',', '.', regex=False)  # Reemplazar comas decimales por puntos

# Convertir las columnas a tipo float primero
for column in columns_to_convert:
    datos[column] = pd.to_numeric(datos[column], errors='coerce')  # Convertir a float, coercing errors to NaN

# Función para convertir a int con manejo de excepciones y relleno de NaN con guion -
def safe_convert_to_int(value):
    try:
        if pd.isnull(value):
            return '-'
        return int(value)
    except (ValueError, TypeError):
        return '-'

# Aplicar la función personalizada a las columnas
for column in columns_to_convert:
    datos[column] = datos[column].apply(safe_convert_to_int)

datos
```

```python
datos.columns
```

```python
columns_to_fill = ['Tasas_Incidencias', 'Confirmados_Laboratorio', '%_Laboratorio_Confirmado', 
                    '%_Dengue_Grave', 'Poblacion_x_1000', 'ID_Serotipo']

# Reemplazar valores vacíos y espacios en blanco con NaN
datos[columns_to_fill] = datos[columns_to_fill].replace({'': np.nan, ' ': np.nan})

# Rellenar los NaN con guiones en las columnas especificadas
datos[columns_to_fill] = datos[columns_to_fill].fillna('-')

# Mostrar el DataFrame actualizado
datos

```

```python
datos
```

```python
datos.to_csv("casos_dengue_final.csv")
```
